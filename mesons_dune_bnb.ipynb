```python
log_ebins = np.logspace(-4, 2, 70)
ebins = np.linspace(0.0, 105.0, 70)
plt.hist(combined_pi0_4vectors[:,3], bins=log_ebins, histtype='step', label=r"$\pi^0$ ($N = 1.19$ / POT)")
plt.hist(combined_eta_4vectors[:,3], bins=log_ebins, histtype='step', label=r"$\eta^0$ ($N = 0.15$ / POT)")
plt.ylabel(r"Mesons / $10^6$ POT", fontsize=14)
plt.xlabel(r"Kinetic Energy [GeV]", fontsize=14)
plt.yscale('log')
plt.xscale('log')
plt.legend(fontsize=12, framealpha=1.0, loc="upper left")
plt.tight_layout()
plt.show()
```

```python
# get the angles wrt lab frame
pi0_theta = np.arccos(combined_pi0_4vectors[:,2] / np.sqrt(combined_pi0_4vectors[:,0]**2 + combined_pi0_4vectors[:,1]**2 + combined_pi0_4vectors[:,2]**2))
eta_theta = np.arccos(combined_eta_4vectors[:,2] / np.sqrt(combined_eta_4vectors[:,0]**2 + combined_eta_4vectors[:,1]**2 + combined_eta_4vectors[:,2]**2))

log_ebins = np.logspace(-3.2, 2.1, 70)
theta_bins = np.logspace(-3, np.log10(np.pi), 50)


plt.hist2d(combined_pi0_4vectors[:,3], pi0_theta, bins=[log_ebins, theta_bins], norm=LogNorm())
plt.ylabel(r"$\theta_{\pi^0}$ [rad]", fontsize=14)
plt.xlabel(r"Kinetic Energy [GeV]", fontsize=14)
plt.colorbar(label=r"$\pi^0$ / $10^6$ POT")
plt.yscale('log')
plt.xscale('log')
plt.tight_layout()
plt.show()

plt.hist2d(combined_eta_4vectors[:,3], eta_theta, bins=[log_ebins, theta_bins], norm=LogNorm())
plt.ylabel(r"$\theta_{\eta^0}$ [rad]", fontsize=14)
plt.xlabel(r"Kinetic Energy [GeV]", fontsize=14)
plt.colorbar(label=r"$\eta$ / $10^6$ POT")
plt.yscale('log')
plt.xscale('log')
plt.tight_layout()
plt.show()
```

```python
theta_bins = np.logspace(-3.5, np.log10(np.pi), 100)

theta_vals = (theta_bins[1:] + theta_bins[:-1])/2
bin_widths = (theta_bins[1:] - theta_bins[:-1])

plt.hist(pi0_theta, bins=theta_bins, histtype='step', label=r"$\pi^0$ ($N = 1.19$ / POT)")
plt.hist(eta_theta, bins=theta_bins, histtype='step', label=r"$\eta^0$ ($N = 0.15$ / POT)")
plt.plot(theta_vals, 1e5 * sin(theta_vals) * bin_widths, label="Isotropic")
plt.ylabel(r"Mesons / $10^6$ POT", fontsize=14)
plt.xlabel(r"Angle", fontsize=14)
plt.yscale('log')
plt.xscale('log')
plt.xlim((theta_bins[0], theta_bins[-1]))
plt.ylim(bottom=1e-1)
plt.legend(fontsize=12, framealpha=1.0, loc="upper left")
plt.tight_layout()
plt.show()
```

### Check how many pi0 / eta are left after angular cut: DUNE

```python
# get the angles wrt lab frame
pi0_theta = np.arccos(combined_pi0_4vectors[:,2] / np.sqrt(combined_pi0_4vectors[:,0]**2 + combined_pi0_4vectors[:,1]**2 + combined_pi0_4vectors[:,2]**2))
eta_theta = np.arccos(combined_eta_4vectors[:,2] / np.sqrt(combined_eta_4vectors[:,0]**2 + combined_eta_4vectors[:,1]**2 + combined_eta_4vectors[:,2]**2))

log_ebins = np.logspace(-3.2, 2.1, 70)
theta_bins = np.logspace(-3, np.log10(np.pi), 50)

# apply cuts for DUNE: arctan(3/574) = 5 mrad

pi0_theta_cut = pi0_theta[pi0_theta < 5e-3]
pi0_T_cut = combined_pi0_4vectors[:,3][pi0_theta < 5e-3]

eta_theta_cut = eta_theta[eta_theta < 5e-3]
eta_T_cut = combined_eta_4vectors[:,3][eta_theta < 5e-3]

print("number of pi0 after cut = {} vs. {}".format(pi0_T_cut.shape, combined_pi0_4vectors[:,3].shape[0]))
print("number of pi0 after cut = {} vs. {}".format(eta_T_cut.shape, combined_eta_4vectors[:,3].shape[0]))

plt.hist(combined_pi0_4vectors[:,3], bins=log_ebins, label=r"$\pi^0$", color='r', histtype='step')
plt.hist(combined_eta_4vectors[:,3], bins=log_ebins, label=r"$\eta$", color='b', histtype='step')
plt.hist(pi0_T_cut, bins=log_ebins, label=r"$\pi^0$ after cut", color='r', ls='dashed', histtype='step')
plt.hist(eta_T_cut, bins=log_ebins, label=r"$\eta$ after cut", color='b', ls='dashed', histtype='step')
plt.ylabel(r"Counts", fontsize=14)
plt.xlabel(r"Kinetic Energy [GeV]", fontsize=14)
plt.yscale('log')
plt.xscale('log')
plt.tight_layout()
plt.show()
```
